```python
# 必要なライブラリをimport
from transformers import T5Tokenizer, T5ForConditionalGeneration
import torch

# 利用可能ならGPUを使用する
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
#↑は以下と等価：
# if torch.cuda.is_available():
#   device = torch.device('cuda')
# else:
#   device = torch.device('cpu')
```

```python
# トークナイザを読み込み
tokenizer = T5Tokenizer.from_pretrained("google/flan-t5-base")
# Transformerモデルを読み込み
model = T5ForConditionalGeneration.from_pretrained("google/flan-t5-base").to(device)
```

```python
# プロンプト
input_text = "summarize: I have a cat. Her name is Tama. She came to my house three years ago. She likes to sleep in a sunny place. After school, I often play with her with a ball. Watching her makes me happy."
print("input_text:{}".format(input_text))
# プロンプトをトークン列に変換
input_ids = tokenizer(input_text, return_tensors="pt").input_ids.to(device)
print("input_ids:{}".format(input_ids))
```

```python
# LLMにトークン列を入力して出力を得る
outputs = model.generate(input_ids)
output_ids = outputs[0]
print("output_ids:{}".format(output_ids))

# 出力トークン列を文字列に変換
output_text = tokenizer.decode(output_ids)
print("output_text:{}".format(output_text))
```
